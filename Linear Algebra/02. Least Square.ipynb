```python
import numpy as np
```

```python
A = np.array([[11,12,13,14], [15,16,17,18]])
print(A)
```

```python
# least squares y ~= ax+b = Ax (bias bÎ•º Ìï©Ïπ®)
# xÎäî ÌïòÎÇòÏùò feature (ÏûÖÎ†•Í∞í) / yÎäî target(Ï∂úÎ†•Í∞í)
# ùê¥tAX = Atb  (Ïó¨Í∏∞ÏÑú bÎäî yÎ•º Í∞ÄÎ¶¨ÌÇ¥)  


x = np.array([0, 1, 2, 3])
y = np.array([-1, 0.2, 0.9, 2.1])
```

```python
# concatnate a constant term

A = np.vstack([x, np.ones(len(x))]).T
A 
```

```python
# create a matrix A * A

At_A = A.T.dot(A)
At_A
```

```python
At_y = A.T.dot(y)
At_y
```

```python
# solve for x

from numpy.linalg import solve

c = solve(At_A, At_y)
c
```

```python
# fitted value
# Í≤∞Í≥ºÍ∞íÏùÄ y Í∞íÍ≥º Í±∞Ïùò ÏùºÏπòÌï¥Ïïº ÌïúÎã§

y_fitted = A.dot(c)
y_fitted
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, 'o', label = 'Original data', markersize=10)
plt.plot(x, y_fitted, 'r', label='Fitted line')
plt.legend()
plt.show()
```

```python
# compute error vector

error = y - y_fitted
error
```

```python
# calculate error  (ÏµúÏÜå error Í∞í / Ïñ¥Îñ§ Ìï¥Î•º Íµ¨Ìï¥ÎèÑ Ïù¥ error Î≥¥Îã§ ÎÇÆÏïÑÏßÄÏßÄ ÏïäÎäîÎã§)

np.sqrt(np.sum(error ** 2))
```

```python
# calculate error using 'norm' function  (Ï£ºÏñ¥ÏßÑ Î≤°ÌÑ∞Ïùò Í∏∏Ïù¥(norm)Î•º Íµ¨Ìï®ÏúºÎ°úÏç® errorÎ•º Íµ¨ÌïúÎã§)

from numpy.linalg import norm

norm(y - y_fitted)
```

# Í∞ïÏùò Ï§ë Ï†úÏãúÎêú Î¨∏Ï†ú ÌíÄÍ∏∞ 

```python
A = np.array([[60, 5.5, 1], [65, 4.8, 0], [55, 6.0, 1], [50, 5.0, 1]])
b = np.array([66, 70,78,72])
```

```python
At_A = A.T.dot(A)
At_A
```

```python
At_b = A.T.dot(b)
At_b
```

```python
r = solve(At_A, At_b)
r
```

```python
# fitted value

b_fitted = A.dot(r)
b_fitted
```

```python
# compute error vector

error = b - b_fitted
error
```

```python
# calculate error

np.sqrt(np.sum(error ** 2))
```

```python
# calculate using 'norn' function

norm(b - b_fitted)
```
