```python
# coefficient matrix A and a vector b
# aT X = b  (aT = A, b =b)

A = np.array([[60, 5.5, 1], [65, 5.0, 0], [55, 6.0, 1]])
b = np.array([66, 70, 78])
```

```python
# identity matrix
eye3 = np.eye(3)
eye3
```

```python
# computing an iverse

from numpy.linalg import inv

A_inv = inv(A)
A_inv
```

```python
# wrong matrix multiplication

A * A_inv
```

```python
# correct matrix multiplication

A.dot(A_inv)
```

```python
# solution of a linear system
# 대부분 역행력을 이용해서 해를 구하지 않는다. 이유는 계산과정에서 소수점이하 오차가 발생하기 때문임 

x = A_inv.dot(b)
x
```

```python
# a better way to solve the dame linear system
# solve는 역행렬이 아닌 방법으로 계산하며, 소수점 오차 발생을 줄일 수 있다. 

from numpy.linalg import solve

x = solve(A, b)
x
```
