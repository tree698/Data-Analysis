```python
import numpy as np
```

```python
# column vector
c = np.array([1,2,3])
print(c.shape)
print(c)

# obtaining a particular entry
print(c[0])
```

```python
# row vector
r = np.array([[1,2,3]])
print(r.shape)
print(r)

# obtaining a particular entry
print(r[0,1])
```

```python
# creating a matrix with all zeros
a = np.zeros((2,2))
print(a)

# creating a matrix with all one
b = np.ones((2,2))
print(b)

# creating a matrix filled with the same constnat
c = np.full((2,2),7)
print(c)

# creating a matrix with random values
d = np.random.random((2,2))
print(d)
```

```python
# creating a matrix
A = np.array([[1,2],[3,4],[5,6]])
print(A)
print(A[0,0])
print(A[0,1])
print(A[1,0])
print(A[1,1])
```

```python
# creating a matrix
B = np.array([[11,12,13,14],[15,16,17,18]])
print(B)
B
```

```python
# matrix-matrix multiplication
np.dot(A,B)
```

```python
# coefficient matrix A and a vector b
# aT X = b  (aT = A, b =b)

A = np.array([[60, 5.5, 1], [65, 5.0, 0], [55, 6.0, 1]])
b = np.array([66, 70, 78])
```

```python
# identity matrix
eye3 = np.eye(3)
eye3
```

```python
# computing an iverse

from numpy.linalg import inv

A_inv = inv(A)
A_inv
```

```python
# wrong matrix multiplication
A * A_inv
```

```python
# correct matrix multiplication
A.dot(A_inv)
```

```python
# solution of a linear system
x = A_inv.dot(b)
x
```

```python
# a better way to solve the dame linear system

from numpy.linalg import solve

x = solve(A, b)
x
```
