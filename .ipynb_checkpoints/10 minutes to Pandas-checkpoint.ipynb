# Missing data

### pandas primarily uses the value np.nan to represent missing data. It is by default not included in computations. 

### Reindexing allows you to change/add/delete the index on a specified axis. This returns a copy of the data:

```python
df1 = df.reindex(index=dates[0:4], columns=list(df.columns) + ["E"])
df1
```

```python
df1.loc[dates[0]:dates[1], "E"] = 1
df1
```

### To drop any rows that have missing data:

```python
df1.dropna(how="any")
```

### Filling missing data:

```python
df1.fillna(value=5)
```

```python
df1.loc[:,["E"]].fillna(value=5)
```

### To get the boolean mask where values are nan:

```python
pd.isna(df1)
```

```python
df1.isna()
```

# Reshaping

## Stack

```python
tuples = list(
    zip(
        *[
            ["bar", "bar", "baz", "foo", "foo", "qux", "qux"],
            ["one", "two", "one", "two", "one", "two", "one"]
        ]))
index = pd.MultiIndex.from_tuples(tuples, names=["first", "second"])
df = pd.DataFrame(np.random.randn(7,2), index=index, columns=["A","B"])
df
```

```python
df2 = df[:4]
df2
```

### The stack() method compresses a leave in the DataFrame's columns

```python
stacked = df2.stack()
stacked
```

### With a stacked DataFrame or Series(having a MultiIndex as the index) the inverse operation of stack(), which by default unstacks the last level:

```python
stacked.unstack()
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(2)
```

## Pivot tables

# Tims series

### pandas has simple, powerful, and efficient functionality for performing resampling operations during frequecny conversion. This is extremely common in, but not limited to, financial applications:

```python
rng = pd.date_range("1/1/2012", periods=100, freq="s")
rng
```

```python
ts = pd.Series(np.random.randint(0, 500, len(rng)), index=rng)
ts
```

```python
ts.resample("5Min").sum()
```

### Time zone representation

```python
rng = pd.date_range("3/6/2012 00:00", periods=5, freq="D")
rng
```

```python
ts = pd.Series(np.random.randn(len(rng)), rng)
ts
```

```python
ts_utc = ts.tz_localize("UTC")
ts_utc
```

### Converting to another time zone

```python
ts_utc.tz_convert("US/Eastern")
```

### Converting between time span representations:

```python
rng = pd.date_range("1/1/2012", periods=5, freq="M")
rng
```

```python
ts = pd.Series(np.random.randn(len(rng)), index=rng)
ts
```

```python
ps = ts.to_period()
ps
```

```python
ps.to_timestamp()
```

### Converting between period and timestamp enables some conventient arithmetic functions to be used. In the following example, we convert a quarterly frequency with year ending in November to 9am of the end of the month following the quarter end:

```python
prng = pd.period_range("1990Q1", "2000Q4", freq="Q-NOV")
prng
```

```python
ts = pd.Series(np.random.randn(len(prng)), prng)
ts
```

```python
ts.index = (prng.asfreq("M", "e") + 1).asfreq("H", "s") +9
ts.head()
```

# Categoricals

### pandas can include categorical data in a DataFrame. 

```python
df = pd.DataFrame(
    {
        "id":[1,2,3,4,5,6],
        "raw_grade":["a", "b", "b", "a", "a", "e"]
    })
df
```

### Convert the raw grades to a categorical data type:

```python
df["grade"] = df["raw_grade"].astype("category")
df["grade"]
```

### Rename the categories to more meaningful names:

```python
df["grade"].cat.categories = ["Very good", "good", "Very bad"]
```

### Reorder the categories and simultaneously add the missing categories:

```python
df["grade"] = df["grade"].cat.set_categories(
                ["very bad", "bad", "medium", "good", "very good"])
df["grade"]
```

### Sorting is per oder in the categories, not lexical order:

```python
df.sort_values(by="grade")
```

### Grouping by a categorical column also shows empty categories

```python
df.groupby("grade").size()
```
