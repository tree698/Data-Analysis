* TF 2.0에서는 오페레이션을 실행하는 순간 연산이 즉시 수행되기 때문에 오퍼레이션 실행 결과를 numpy() 메서드를 통하여 바로 알수 있음

```python
a = tf.constant(10)
b = tf.constant(20)

c = a + b
d = (a+b).numpy()

print(type(c))
print(c)
print(type(d), d)

d_numpy_to_tensor = tf.convert_to_tensor(d)

print(type(d_numpy_to_tensor))
print(d_numpy_to_tensor)
```

* TF 2.0에서는 tf.Variable()값을 초기화하기 위해 세션 내에서 tf.global_variables_initializer() 과정이 필요 없으며 변수를 정의함과 동시에 초깃값이 할당됨 

```python
# 가우시안 분포
W = tf.Variable(tf.random.normal([1]))
print('initial W = ', W.numpy())
print('==============')

for step in range(3):
    W = W + 1.0
    print('step = ', step, ', W = ', W.numpy())
```

```python
a = tf.constant(1.0)
b = tf.constant(3.0)

# 함수 정의
def tensor_sum(x, y):
    return x + y

result = tensor_sum(a, b)

print(type(result))
print(result.numpy())
```
