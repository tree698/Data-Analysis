```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD, Adam

import numpy as np

print(tf.__version__)
```

```python
x_data = np.array([[1,2,0], [5,4,3], [1,2,-1], [3,1,0], [2,4,2],
                  [4,1,2], [-1,3,2],[4,3,3],[0,2,6],[2,2,2],
                  [1,-2,-2], [0,1,3], [1,1,3], [0,1,4], [2,3,3]])
y_data = np.array([-4,4,-6,3,-4,
                  9,-7,5,6,0,
                  4,3,5,5,1])

print('x_data.shpae = ', x_data.shape, 'y_data.shape = ', y_data.shape)
```

# 모델 구축

```python
model = Sequential()
model.add(Dense(1, input_shape=(3,), activation = 'linear'))
```

# 모델 컴파일

* SGD(Stochastic Gradient Descent) : 학습 알고리즘
* loss : 손실함수
* mse : mean square error (평균제곱오차)

```python
model.compile(optimizer = SGD(learning_rate=1e-2), loss='mse')
model.summary()
```

# 모델 학습

```python
hist = model.fit(x_data, y_data, epochs=1000)
```

# 모델 평가 및 예측

```python
test_data = [[5,5,0],[2,3,1],[-1,0,-1], [10,5,2],[4,-1,-2]]

ret_val = [2*data[0] - 3*data[1] + 2*data[2] for data in test_data]

prediction_val = model.predict(np.array(test_data))

print(prediction_val)
print("====================")
print(ret_val)
```

# 모델 가중치, 바이어스

```python
print(model.input)
print(model.output)
print(model.weights)
```

# 모델 손실함수 추세

```python
import matplotlib.pyplot as plt

plt.title('Loss Trend')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid()

plt.plot(hist.history['loss'], label='train loss')
plt.legend(loc='best')

plt.show()
```
